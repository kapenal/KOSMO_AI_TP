```python
import requests

# 위도
lat = { 
    "서울" : "37.5665",
    "인천" : "37.503318892125",
    "강릉" : "37.75324012",
    "대전" : "36.3577577",
    "광주" : "35.1500822",
    "부산" : "35.1731121",
    "제주" : "33.3785614"
}
# 경도
lon = {
    "서울" : "126.9780",
    "인천" : "126.66566350799",
    "강릉" : "128.8753382",
    "대전" : "127.3867458",
    "광주" : "126.8559071",
    "부산" : "129.0714122",
    "제주" : "126.5661908"
}

# api
site_url = "https://api.openweathermap.org/data/2.5/weather?"
API_key = "c45f08438d648b1290fe15b338cad6eb"
UNIT = "metric"
weather_data = []  # 도시별 날씨 정보를 저장할 리스트


# 도시별로 호출
for city in lat.keys():
    parameter = f"lat={lat[city]}&lon={lon[city]}&appid={API_key}&units={UNIT}"
    url_para = site_url + parameter
    
    # url에 get 요청을 보내고 결과를 변수에 저장, 성공하면 200 반환
    response = requests.get(url_para)
    # 응답 상태 확인
    print(f"[HTTP 응답 상태코드] = [{response.status_code}]")
    
    if response.status_code == 200:
        # json 응답을 Python 딕셔너리로 파싱
        data = response.json()

        # 데이터 가지고 오기
        weather = data["weather"][0]["description"] # 날씨
        wtd_icon = data["weather"][0]["icon"] # 아이콘
        temp = data["main"]["temp"] # 온도
        
        # 라스트화
        weather_data.append({
            "도시": city,
            "기온": temp,
            "날씨": weather,
            "이미지" : wtd_icon
        })
    else:
        print(f"{city}: 날씨 정보를 가져오지 못했습니다. (에러코드: {response.status_code})\n")
        print(f'API 요청 실패. (오류 메세지 확인 :{response.text})\n')

print(f"\n날씨 데이터 = {weather_data}\n")

# 결과 출력
for i in weather_data:
    print(f"도시: {i['도시']}")
    print(f"기온: {i['기온']}°C")
    print(f"날씨: {i['날씨']}")
    print(f"아이콘: {i['이미지']}")
```
